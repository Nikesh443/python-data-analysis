# series

```python
s=pd.Series([1,2,3,4,5])
print(s)
print(type(s))

# index
print(s[0])
print(s[2])
```

```python
s = pd.Series([23,24,25,26],index=["Adam","Eve","Bob","Charlie"])
s
# These names are actually named index(labels)
```

```python
print(s["Adam"]) 
print(s["Eve"])
```

```python
print(s.index)
```

```python
# vectorized operations
s1 = pd.Series([11,12,34,45])
s2 = pd.Series([11,12,35,45])
print(s1 + s2)

s1[0]=100
print(s1[0])

changed_s1 = s1.drop(0)
print(s1)
print(changed_s1)
```

# DataFrame

```python
info = {
    "Name":['Adam',"Eve","Bob"],
    "Age":[23,24,25],
    "GPA":[9.5,8.5,7.2]
}
df=pd.DataFrame(info)
print(df)
```

```python
print(df.index)
```

```python
print(df.columns)
```

```python
# lists of lists
df = pd.DataFrame([["Adam",23],["Nikesh",22],["Eve",23]],columns=["Name","Age"])
```

```python
print(df)
```
