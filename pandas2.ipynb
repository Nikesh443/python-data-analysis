# cleaning data

```python
df2 = pd.read_csv("raw_data.csv")
df2
```

```python
# isnull(),isna()
df2.isnull()
df2.isna()
df2.isna().sum()
```

```python
df2.dropna(inplace=True) # drop rows
df2.dropna(axis=1,inplace=True) # drop columns
df2.isna().sum()
```

```python
# using average for fillna() for age column
cleaned_df = df2.copy()
age_mean = cleaned_df["age"].mean()
cleaned_df["age"]=df2["age"].fillna(age_mean)
cleaned_df
```

```python
# ffill
ffill_df = df2.copy()
ffill_df.ffill()
```

```python
#bfill()
bfill_df = df2.copy()
bfill_df.bfill()
```
